```python
%%bash
source /usr/local/etc/profile.d/conda.sh
conda activate atlasnet

obj="airplane"
gt_folder="$GOOGLE_DRIVE_PATH/${obj}/GT"

at_folder="$GOOGLE_DRIVE_PATH/${obj}/AT"
mkdir -p "$at_folder"

for gt_file in "$gt_folder"/*.obj; do
    filename=$(basename -- "$gt_file")
    base_filename="${filename%.*}"

    python train.py --demo --demo_input_path "$gt_file" # --reload_model_path ./training/trained_models/atlasnet_AE_25_patches.pth # ./training/atlasnet_singleview_1_sphere/network.pth

    output_file="${base_filename}AtlasnetReconstruction.obj"
    mv "$gt_folder/$output_file" "$at_folder/"
done
```
